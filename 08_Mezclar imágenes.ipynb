```python
imagen_1 = cv2.imread('perro.jpg')
imagen_1 = cv2.cvtColor(imagen_1, cv2.COLOR_BGR2RGB)
plt.imshow(imagen_1)
```

```python
imagen_2 = cv2.imread('marca-agua-redaitcurso.jpg')
imagen_2 = cv2.cvtColor(imagen_2, cv2.COLOR_BGR2RGB)
plt.imshow(imagen_2)
```

```python
imagen_1.shape
```

```python
imagen_2.shape
```

```python
imagen_1 = cv2.resize(imagen_1,(1200,1200))
```

```python
imagen_2 = cv2.resize(imagen_2,(1200,1200))
```

```python
plt.imshow(imagen_1)
```

```python
plt.imshow(imagen_2)
```

```python
imagen_mezcla = cv2.addWeighted(src1 = imagen_1, alpha = 0.5, src2 = imagen_2, beta = 0.5, gamma = 0)
```

```python
plt.imshow(imagen_mezcla)
```

```python
imagen_mezcla = cv2.addWeighted(src1 = imagen_1, alpha = 0.8, src2 = imagen_2, beta = 0.2, gamma = 0)
plt.imshow(imagen_mezcla)
```
