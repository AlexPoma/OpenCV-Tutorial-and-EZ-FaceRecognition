```python
imagen = cv2.imread('perro.jpg')
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen)
```

```python
imagen.shape
```

```python
imagen_2 = cv2.resize(imagen, (600,500))
```

```python
plt.imshow(imagen_2)
```

```python
ratio_ancho = 0.5
ratio_alto = 0.5
```

```python
imagen_2 = cv2.resize(imagen_2,(0,0),imagen_2,ratio_ancho,ratio_alto)
```

```python
plt.imshow(imagen_2)
```

```python
imagen_2.shape
```

```python
imagen_2 = cv2.flip(imagen_2,1)
```

```python
plt.imshow(imagen_2)
```

```python
imagen_2 = cv2.flip(imagen_2,0)
```

```python
plt.imshow(imagen_2)
```

```python
imagen_2 = cv2.cvtColor(imagen_2, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(imagen_2)
```

```python
cv2.imwrite('perro_al_reves.jpg',imagen_2)
```
